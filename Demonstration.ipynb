```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import scipy.stats as st
from Plotting_Functions import intersection, plotline, findyinter, romina
from Generate_data_dict import generate_data_dict


path="Resources/US_Data.csv"
tuition=[0, -96464, 2, -166464, 4, -96464]
```

```python
message1='Welcome to use our program Does Education Pay Off.' 
message2='This program will generate you a custom plot of the accumulate income of each education level of undergrad, master, and Phd for a field of your choice.' 
message3='From this plot we will see the likelihood of whether your graduate education will eventually pay off, or and how long that will likely take.'
message4='Please choose a major from the above selection, insert "all" to see all majors: '
majors=['Computer Science',
        'Mathematics' ,
        'Statistics and Decision Science', 
        'Chemical Engineering',
        'Civil Engineering',
        'Mechanical Engineering', 
        'Biology', 
        'Environmental', 
        'Chemistry', 
        'Economics', 
        'Political Science', 
        'Sociology', 
        'General Education', 
        'Law', 
        'Construction Services', 
        'Accounting', 
        'Finance', 
        'Marketing and Marketing Research', 
        'Pharmacy, Pharmaceutical Sciences and Administration',
        'Nursing', 
        'General Medical Health Services']

str_build_majors = ''
for i in majors:
    str_build_majors += f'{i}\n'

str_build = f'{message1} \n\n{message2} \n\n{message3} \n\n{str_build_majors}'
print(str_build)

chosen_major = input(message4)

upper_cut_off =float(input("Please eneter upper income cutoff percentile for source data, 85% recommended at least: "))
upper_cut_off = upper_cut_off/100

lower_cut_off = float(input("Please eneter lower income cutoff percentile for source data, 15% recommended at least: " ))
lower_cut_off = lower_cut_off/100

percentile_ug = float(input("Please enter income percentile of undergrad you'd like to compare with: "))
percentile_ug = percentile_ug/100


```

```python
dictdata=generate_data_dict(path, upper_cut_off, lower_cut_off)
majornames,incomes=romina(dictdata)
```

```python
if chosen_major=='all':

    for i in range(len(majornames)):
        major=majornames[i]
        income=incomes[i]
        intercept=findyinter(income, tuition)
        plotline(income,intercept,major, percentile_ug)
        
else:

    i=majornames.index(chosen_major)
    major=majornames[i]
    income=incomes[i]
    intercept=findyinter(income, tuition)
    plotline(income,intercept,major, percentile_ug)
```
